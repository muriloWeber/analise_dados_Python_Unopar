# DESAFIO

*   Apresentar comparação da quantidade de satélites que são brasileiros dos que são estrangeiros.

*   Dentre os brasileiros, discriminar a quantidade de cada operadora comercial

*   Quantidade de satélites operando em cada banda

*   As mesmas análises para os satélites estrangeiros



 

# RESOLUÇÃO

Fonte de dados:  https://www.dados.gov.br/dataset/relacao-de-satelites-geoestacionarios-autorizados-a-operar-no-brasil


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Obtendo dados através de dataset em .csv e transformando em DataFrame Pandas
df_satelites = pd.read_csv('satelites_operando_comercialmente.csv', sep=';')
# Dropando linhas duplicadas do dataset
df_satelites.drop_duplicates(inplace=True)
# Com as linhas duplicadas dropadas é necessário resetar o índice do dataframe
df_satelites.reset_index(drop=True, inplace=True)

print(df_satelites.info())
df_satelites.head()
```

```python
# Quantos satélites são brasileiros e quantos são estrangeiros?

# Criando gráfico para responder a pergunta de negócio
plt.figure(figsize=(7,3))
plt.tick_params(labelsize=12)
plt.title('Satélites Brasileiros X Estrangeiros')
ax = sns.countplot(data=df_satelites, x='Tipo de Direito')
ax.set_ylabel('Quantidade')
sns.set_theme(style="darkgrid")

# iterar pelos patches do gráfico para imprimir os rótulos
for i in ax.patches:
    ax.annotate(i.get_height(),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center',
                  va='baseline',
                  fontsize=12,
                  color='black',
                  xytext=(0, 1),
                  textcoords='offset points')
```

```python
# quantos satétiles cada operadora brasileira possui operando?

# criando novo dataframe com filtro de apenas satélites brasileiros
df_satelites_brasileiros = df_satelites.loc[df_satelites['Tipo de Direito'] == 'Brasileiro']

plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
plt.tick_params(labelsize=12)
plt.title('Satélites Brasileiros por Operadora')
ax1 = sns.countplot(data=df_satelites_brasileiros, x='Operadora comercial')
ax1.set_ylabel('Quantidade')
sns.set_theme(style="darkgrid")
for i in ax1.patches:
    ax1.annotate(i.get_height(),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center',
                  va='baseline',
                  fontsize=12,
                  color='black',
                  xytext=(0, 1),
                  textcoords='offset points')
```

```python
# Quantos satélites brasileiros estão operando em cada banda?

plt.figure(figsize=(15,5))
plt.xticks(rotation=90)
plt.tick_params(labelsize=12)
plt.title('Satélites Brasileiros por Banda')
ax2 = sns.countplot(data=df_satelites_brasileiros, x='Bandas do satélite')
ax2.set_ylabel('Quantidade')
sns.set_theme(style="darkgrid")
for i in ax2.patches:
    ax2.annotate(i.get_height(),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center',
                  va='baseline',
                  fontsize=12,
                  color='black',
                  xytext=(0, 1),
                  textcoords='offset points')
```

```python
# Quantos satétiles cada operadora estrangeira possui operando?

#criando novo df com apenas satélites estrangeiros
df_satelites_estrangeiros = df_satelites.loc[df_satelites['Tipo de Direito'] == 'Estrangeiro']

plt.figure(figsize=(15,5))
plt.xticks(rotation=90)
plt.tick_params(labelsize=12)
plt.title('Satélites Estrangeiros por Operadora')
ax3 = sns.countplot(data=df_satelites_estrangeiros, x='Operadora comercial')
ax3.set_ylabel('Quantidade')
sns.set_theme(style='darkgrid')
for i in ax3.patches:
    ax3.annotate(i.get_height(),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center',
                  va='baseline',
                  fontsize=12,
                  color='black',
                  xytext=(0, 1),
                  textcoords='offset points')
```

```python
# Quantos satélites estrangeiros estão operando em cada banda?

plt.figure(figsize=(15,5))
plt.xticks(rotation=90)
plt.tick_params(labelsize=12)
plt.title('Satélites Estrangeiros por Banda')
ax4 = sns.countplot(data=df_satelites_estrangeiros, x='Bandas do satélite')
ax4.set_ylabel('Quantidade')
sns.set_theme(style='darkgrid')
for i in ax4.patches:
    ax4.annotate(i.get_height(),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center',
                  va='baseline',
                  fontsize=12,
                  color='black',
                  xytext=(0, 1),
                  textcoords='offset points')
```
